## 💼 Scenario 3: Portfolio-Aware News Prioritisation
**Judge Scenario:** RBI rate cut + IT regulation — which matters MORE for THIS investor's portfolio?

```python
# ─── CELL 9: Run Scenario 3 ──────────────────────────────────
result_s3 = agent.run('portfolio')
agent.print_report(result_s3)
```

```python
# ─── CELL 10: View P&L impact breakdown ──────────────────────
for event in result_s3.get('impact_analysis', []):
    print(f"\n{event['event_title']}")
    print(f"  Total impact: {event['impact_direction']} ₹{event['total_portfolio_impact_rupees']:,.0f} ({event['total_portfolio_impact_pct']:.2f}%)")
    for holding in event.get('affected_holdings', []):
        sign = '+' if holding['direction'] == 'GAIN' else '-'
        print(f"  {holding['stock_name']:20} {sign}₹{abs(holding['impact_rupees']):,.0f}")
```
